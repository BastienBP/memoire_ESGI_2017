```python
from kafka import KafkaConsumer
from kafka import TopicPartition
import model_rf_fraud_detection as rf
import pandas as pd
from pandas import Index
import numpy as np
```

```python
def set_model():
    clf, df = rf.initialize_model()
    columns = Index([u'V1', u'V2', u'V3', u'V4', u'V5', u'V6', u'V7', u'V8', u'V9', u'V10',
       u'V11', u'V12', u'V13', u'V14', u'V15', u'V16', u'V17', u'V18', u'V19',
       u'V20', u'V21', u'V22', u'V23', u'V24', u'V25', u'V26', u'V27', u'V28',
       u'Amount'],
      dtype='object')
    df_test = pd.DataFrame(columns=columns)
    return clf, df, df_test

def initialize_kafka():
    topic = TopicPartition('foobar', 0)
    consumer = KafkaConsumer(bootstrap_servers='localhost:9092')
    consumer.assign([topic])
    #consumer.seek_to_beginning(topic)
    
    return consumer

def determine_type(value):
    if value == 0:
        print "Normal"

    elif value == 1:
        print "Fraud!"
```

```python
clf, df_test, df_stream = set_model()
```

```python
consumer = initialize_kafka()
list_res = []
for msg in consumer:
    row = eval(msg.value)
    df_stream.loc[0] = row
    value = rf.load_model(clf, df_stream)[0]
    determine_type(value)
    list_res.append(value)
```
